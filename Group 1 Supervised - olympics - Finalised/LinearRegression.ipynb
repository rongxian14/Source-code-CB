## Discretization Method

```python
import pandas as pd

# Example DataFrame with a continuous variable 'Value' based on athlete agaes from 2012 to 2016
data = pd.DataFrame({'Value' : [18 , 19 , 20 , 21 , 23 , 24 , 25 , 26 , 27 , 28 , 30 , 31 , 34 , 35 , 36 , 37]})

# Define bin edges
bins = [10 , 20 , 30 , 40]

# Optional : Define bin labels
labels = ['Low', 'Medium' , 'High']

# Create a new column 'Category' to store the bin labels
data['Category'] = pd.cut(data['Value'], bins=bins, labels=labels, include_lowest=True)

print(data)
```

## EVALUATION

```python
import matplotlib.pyplot as plt

# Plot original continuous variable
plt.hist(data['Value'], bins='auto', alpha=0.5, label='Original')

# Display a bar chart for the discretized variable
plt.bar(data['Category'].value_counts().index, data['Category'].value_counts(), alpha=0.5, label='Discretized')

plt.legend()
plt.show()

# Display descriptive statistics
print(data[['Value', 'Category']].describe())
```

```python
eliminated_features = ["age","height","weight"]
teams.drop(columns=eliminated_features, axis=1)
```

```python
teams = teams[["team", "country", "year", "athletes","events", "prev_medals", "medals"]]
```

# Supervised Learning

## Data spliting

```python
train = teams[teams["year"] < 2012].copy()
test = teams[teams["year"] >= 2012].copy()
# Only used 2012 and 2016
# Choose data from the past so that we can train our machine learning model we want
# Make sure we dont use future data to predict the past
```

```python
train.shape
```

```python
test.shape
```

```python
predictors = ["athletes", "prev_medals"]
```

```python
X_train = train[predictors]
X_train
```

```python
X_test = test[predictors]
X_test
```

```python
y_train = train["medals"]
y_test = test["medals"]
```

# Model building

## Linear Regression

### Model Training

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

lr_results
```

## Prediction on the USA team

```python
test["predictions"] = y_lr_test_pred.round()
```

```python
test[test["team"] == "USA"]
```

## Optimization

### Data Splitting

```python
predictors = ["athletes", "prev_medals", "events"]
```

```python
X_train2= train[predictors]
X_train2
```

```python
X_test2 = test[predictors]
X_test2
```

```python
y_train2 = train["medals"]
y_test2 = test["medals"]
```

### Model Building

```python
from sklearn.linear_model import LinearRegression

lr2 = LinearRegression()
lr2.fit(X_train2, y_train2)
```

```python
y_lr2_train_pred = lr2.predict(X_train2)
y_lr2_test_pred = lr2.predict(X_test2)
```

### Evaluate performance

```python
from sklearn.metrics import mean_squared_error, r2_score

lr2_train_mse = mean_squared_error(y_train2, y_lr2_train_pred)
lr2_train_r2 = r2_score(y_train2, y_lr2_train_pred)

lr2_test_mse = mean_squared_error(y_test2, y_lr2_test_pred)
lr2_test_r2 = r2_score(y_test2, y_lr2_test_pred)
```

```python
lr2_results = pd.DataFrame(['Optimization', lr2_train_mse, lr2_train_r2, lr2_test_mse, lr2_test_r2]).transpose()
lr2_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

lr2_results
```

## Prediction on the USA team

```python
test["predictions"] = y_lr2_test_pred.round()
```

```python
test[test["team"] == "USA"]
```
